```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver
import requests
from urllib.request import urlopen as uReq
```

```python
url_gaana = "https://gaana.com/search/albums/"
keyword = 'trap'
url_gaana = url_gaana + keyword
```

```python
driver = webdriver.Chrome()
driver.get(url_gaana)
source = driver.page_source
gaana_html = bs(source, 'html.parser')
```

```python
boxes = gaana_html.findAll('div',{'class':'card_layout_data'})
```

```python
for a in range(10):
    box = boxes[a]
    album_link = "https://gaana.com" + box.div.a['href']
    
    album_Client = uReq(album_link)
    album_page = album_Client.read()
    album_html = bs(album_page,'html.parser')
    
    album_box = album_html.find('div',{'class':'_t1'})
    album_name = album_box.h1.text
    print('Name of the album: ', album_name)
    
    track_boxes = album_html.findAll('ul',{'class':'s_l artworkload'})
    print('No. of tracks: ', len(track_boxes),'\n')
    
    titles = []
    artists = []
    length = []
    track_list = []
    
    for track in track_boxes:
        track_title = track.find('div',{'class':'track_npqitemdetail'}).span.text
        titles.append(track_title)

        track_artist = track.find('div',{'class':'track_npqitemdetail'}).p.text.split('-')[-1]
        artists.append(track_artist)

        track_length = track.find('li',{'class':'s_duration'}).a.text
        length.append(track_length)

        track_list.append((track_title, track_artist, track_length))
        
    album = pd.DataFrame({'Track Title': titles,'Artists':artists,'Duration':length})
    print(album)
    print('\n \n')
```
